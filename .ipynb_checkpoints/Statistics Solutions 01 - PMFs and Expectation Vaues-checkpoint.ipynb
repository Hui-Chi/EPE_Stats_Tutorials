## Exercise 1: Plot Gaussian, Poisson, binomial distribution based on some given PMFs

Below are the PMFs for three different distributions. Note that the PMF is the discrete counterpart of the continuous PDF. Create a function for each distribution, and plot them using matplotlib!

Probability mass function (PMF) for the normal distribution:

## $$g(x|\mu,\sigma) = \frac{1}{\sqrt{2\pi\,\sigma^2}}exp[-\frac{(x-\mu)^2}{2\,\sigma^2}]$$


Probability mass function (PMF) for the poisson distribution: 

## $$p(x|\mu) = exp[-\mu]\frac{\mu^x}{x!}\text{ for }x\geq0$$


Probability mass function (PMF) for the binomial distribution: 

## $$p(x| N) = \frac{N!}{x!(N-x)!}\frac{1}{2^N}\,\,\,, x\in \text{N}$$

 

```python
def Gaussian(x,mu,sigma):
    return 1/math.sqrt(2*math.pi*sigma**2)*math.exp(-(x-mu)**2 / (2*sigma**2))

def Poisson(x,mu):
    return math.exp(-mu) / math.factorial(x) * mu**x  # x must be an integer >= 0

def Binomial(x,N):
    return math.factorial(N)/(math.factorial(x)*math.factorial(N-x)*2**N) # where x is less than N, and also an integer

lim = 100

# Gaussian
xvals = np.arange(0,xlim)
# note that np.arange includes 0 and excludes xlim=100. There are xlim=100 values, 0 thru 99.
mu = 20
sigma = 5
y = np.vectorize(Gaussian)(xvals,mu,sigma)
plt.plot(xvals,y)
plt.show()

# Poisson
xvals = np.arange(0,xlim)
mu = 20
y = np.vectorize(Poisson)(xvals,mu)
plt.plot(xvals,y)
plt.show()

# Binomial
xvals = np.arange(0,xlim) 
N = 20
y = np.vectorize(Poisson)(xvals,N)
plt.plot(xvals,y)
plt.show()

# Note that I picked values that made the distributions all similar. Isn't their coincidence interesting?
```

## Exercise 2: Plot the PMFs as binned histograms instead of continuous distributions
Note: it can help if you think about the binned histogram value as the expectation value for that bin! You can use numpy.histogram.

```python
# Gaussian
xvals = np.arange(0,100)
mu = 20
sigma = 5
yG = np.vectorize(Gaussian)(xvals,mu,sigma)

# Poisson
xvals = np.arange(0,100)
mu = 20
yP = np.vectorize(Poisson)(xvals,mu)


# Binomial
xvals = np.arange(0,100)
N = 20
yB = np.vectorize(Poisson)(xvals,N)


yG = np.ndarray.tolist(yG)
yP = np.ndarray.tolist(yP)
yB = np.ndarray.tolist(yB)


# Apologies, I referenced the wrong function! What you want is to get the frequencies for each bin, and then plot
# those using a bar chart.

# we choose 50 bins, each bin is 2 wide. 
# Get expectation value for each bin:
expecG = [0]*50
expecP = [0]*50
expecB = [0]*50

for index in np.arange(0,98,2):
    expecG[int(index/2)]=np.trapz([yG[index], yG[index+1], yG[index+2]])
    expecP[int(index/2)]=np.trapz([yP[index], yP[index+1], yP[index+2]])
    expecB[int(index/2)]=np.trapz([yB[index], yB[index+1], yB[index+2]])

lower = np.ndarray.tolist(np.arange(0,100,2))
# print(len(lower))
    
# plot expectation value in each bin 
fig, ax = plt.subplots()
ax.bar(lower, expecG, width=2)
plt.show()

fig, ax = plt.subplots()
ax.bar(lower, expecP, width=2)
plt.show()

fig, ax = plt.subplots()
ax.bar(lower, expecB, width=2)
plt.show()
```

## Exercise 3: Randomly sample the PMFs

Generate 10,000 values such that the histogram of the values follows a gaussian/poisson/binomial distribution. Randomly picking one such value according to a distribution is known as "sampling" that distribution. This is effectively what monte-carlo simulations are doing behind the scenes, except they sample a *lot* of complex distributions to approximate the overall behavior of some variable. Try using random.uniform(0,1) as one of your inputs to your pmf, or alternatively, try the hit-or-miss method on page 17 of these slides! https://hep1.phys.ntu.edu.tw/~kfjack/lecture/hepstat/in2/inter-2.pdf

```python

# With scipy.poisson, we can do this super easily:
# def generate_nm(nexpec,mexpec,numev):
#     ngen = [poisson.ppf(random.uniform(0,1),nexpec) for i in range(0,numev)]
#     mgen = [poisson.ppf(random.uniform(0,1),mexpec) for i in range(0,numev)]
#     return [ngen, mgen]


# Building it ourselves:

# Hit or miss method:
# generate N random x and y values. 
# to generate random numbers: random.choice(listofpotentialelements)
def gen(N):
    xvals = [0]*N
    yvals = [0]*N
    for index in range(0,N):
        xvals[index] = random.choice(np.ndarray.tolist(np.arange(0,100,1))) 
        # we have to use a step size of one so we can use it as input to the poisson PDF, otherwise the 
        # factorial won't work out
        yvals[index] = random.choice(np.ndarray.tolist(np.arange(0,0.3,0.001)))
    return [xvals, yvals]

numevents = 10000000
G_xvals = []
P_xvals = []
B_xvals = []

# Generate some random numbers
[xvals, yvals] = gen(numevents)
for event in range(0,numevents):
    # if the x/y combination passes the selection, we save the x-value
    if yvals[event] < Gaussian(xvals[event],mu,sigma):
        G_xvals.append(xvals[event])
    if yvals[event] < Poisson(xvals[event],mu):
        P_xvals.append(xvals[event])
#     if yvals[event] < Binomial(xvals[event],N):
#         B_xvals.append(xvals[event])

# we plot a density histogram of the x-values, let the code determine the appropriate number of bins
plt.hist(G_xvals,bins=40,range=(0,100),density=True,histtype='step')
plt.show()

plt.hist(P_xvals,bins=40,range=(0,100),density=True,histtype='step')
plt.show()

# Note that if the bin width is too small, you get a really wacky lookin histogram
```
